```python
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
url='http://www.ctaa.com.tn/services-en-ligne/annuaire-des-entreprises/'
driver.get(url)
url='http://www.ctaa.com.tn/services-en-ligne/annuaire-des-entreprises/?tx_wddirectory_pi1%5Bmod%5D='
clients = {'Raison sociale':[], 'Dénomination':[], 'Marques Commerciales':[], 'Produits':[], 'Activités':[], 'Branches':[], 'Régime':[], 'Chiffre d affaires':[], 'Date d entrée en production':[], 'Adresse':[], 'Gouvernorat':[], 'Phone':[], 'Fax':[], 'Web':[], 'Email':[], 'Responsable':[]}

for i in range(975,756,-1) :
  l=[]
  url1=url+str(i)
  driver.get(url1)
  soup=BeautifulSoup(driver.page_source, 'html.parser')
  table = soup.find('table', class_ = 'details')
  for data in table.find_all('tbody') :
    for rows in data.find_all('tr'):
      row=rows.find('td').get_text()
      l.append(row)
   
  if (len(l)==16) :
    print(i,"done")
    clients['Raison sociale'].append(l[0])
    clients['Dénomination'].append(l[1])
    clients['Marques Commerciales'].append(l[2])
    clients['Produits'].append(l[3])
    clients['Activités'].append(l[4])
    clients['Branches'].append(l[5])
    clients['Régime'].append(l[6])
    clients['Chiffre d affaires'].append(l[7])
    clients['Date d entrée en production'].append(l[8])
    clients['Adresse'].append(l[9])
    clients['Gouvernorat'].append(l[10])
    clients['Phone'].append(l[11])
    clients['Fax'].append(l[12])
    clients['Web'].append(l[13])
    clients['Email'].append(l[14])
    clients['Responsable'].append(l[15])
  else :
    print(i,"no")



  
```
